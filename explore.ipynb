```python
fire_aggregated = "data/FIRE/fire-data/fire_aggregated_dataset.json"
fire_msrvtt = "data/FIRE/fire-data/fire_msrvtt_dataset.json"
```

```python
def open_json(path):
    with open(path) as f:
        data = json.load(f)
    return data
```

```python
fire_data = open_json(fire_msrvtt)
```

```python
fire_data.keys()
```

```python
ans = fire_data["annotations"]
```

```python
import pandas as pd
df = pd.DataFrame(ans)
```

```python
df
```

```python
queries = [a["query"] for a in ans]
print("Total queries", len(queries))
print("Unique queries", len(set(queries)))
```

```python
set(queries)
```

```python
df.to_excel("fire_msrvtt_dataset.xlsx")
```

```python
# Take the first 10 queries
unique_queries = list(set(queries))

# Initialize lists to store relevant and non-relevant IDs
relevant_ids = []
non_relevant_ids = []

# Iterate through the first 10 queries and categorize the IDs
for query in unique_queries:
    relevant = []
    non_relevant = []
    for annotation in ans:
        if annotation["query"] == query:
            if annotation["label"] == "relevant":
                relevant.append(annotation["video_id"])
            else:
                non_relevant.append(annotation["video_id"])
    relevant_ids.append(relevant)
    non_relevant_ids.append(non_relevant)

# Create a DataFrame to display the results
result_df = pd.DataFrame({
    "query": unique_queries,
    "relevant_ids": relevant_ids,
    "non_relevant_ids": non_relevant_ids
})

result_df
```

```python
result_df.to_excel("q_judgements.xlsx")
```
